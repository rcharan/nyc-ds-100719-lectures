### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user',delimiter='|')
```

```python
df.head()
```

### Step 4. Discover what is the mean age per occupation

```python
df.groupby('occupation').mean()[['age']].sort_values('age')
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
df['pct_male'] = df['gender'].replace({'M' : 1, 'F': 0})
df.groupby('occupation').mean()[['pct_male']].sort_values('pct_male', ascending = False)
```

```python
df.groupby('occupation').mean()[['pct_male']].sort_values('pct_male', ascending = False)
```

```python
df.query('occupation == "doctor"')
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
df.groupby('occupation')[['age']].aggregate(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
df.groupby(['occupation', 'gender']).mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
g = df.groupby(['occupation', 'gender'])[['user_id']].count().rename({'user_id' : 'pct'}, axis = 1).reset_index()
g['pct'] = g.groupby('occupation').transform(lambda x : x/x.sum())
g
```

```python
def normalize_df(d):
    d['pct'] = d['pct']/d['pct'].sum()
    return d
```
